# Architecture Figure

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1, 2, figsize=(18, 5))

# ==== Figure 1: Memory Complexity vs. Sequence Length ====
L = np.logspace(2, 5, 100)
memory_transformer = L ** 2
memory_linear = L

axes[0].plot(L, memory_transformer, label='Transformer (O(LÂ²))', color='red')
# Custom dash patterns: Mamba dashed with long dashes, RNN dash-dot with short dots
axes[0].plot(L, memory_linear, label='Mamba (O(L))', color='blue', linestyle=(0, (5, 5)))  # dashed: 5 on, 5 off
axes[0].plot(L, memory_linear, label='RNN (O(L))', color='green', linestyle=(0, (3, 1, 1, 1)))  # dash-dot with short dashes/dots
axes[0].set_xscale('log')
axes[0].set_yscale('log')
axes[0].set_xlabel('Sequence Length (L)')
axes[0].set_ylabel('Memory Usage (arb. units)')
axes[0].set_title('Memory Complexity vs. Sequence Length')
axes[0].legend()

# ==== Figure 2: Effective Memory Horizon ====
L = 100
tokens = np.arange(L)
transformer_influence = np.zeros(L)
transformer_influence[-30:] = 1

rnn_influence = np.exp(-0.05 * (L - tokens))
mamba_influence = np.exp(-0.02 * (L - tokens))

axes[1].plot(tokens, transformer_influence, label='Transformer', color='red')
# Add markers and change linewidth for clearer difference
axes[1].plot(tokens, rnn_influence, label='RNN', linestyle='-.', color='green', marker='o', markevery=10, linewidth=2)
axes[1].plot(tokens, mamba_influence, label='Mamba', linestyle='--', color='blue', marker='s', markevery=10, linewidth=2)
axes[1].set_xlabel('Input Token Index')
axes[1].set_ylabel('Effective Influence')
axes[1].set_title('Effective Memory Horizon')
axes[1].legend()

plt.tight_layout()
plt.savefig("mamba-transformer.png")
```
